```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('store_data.csv', encoding_errors='ignore')

df.head()
```

```python
# Profit by State

df_store = df.groupby('State')['Profit'].sum()
df_store
```

```python
# Profit for all the cities of Texas

df_texas = df[df['State'] == 'Texas']
df_tx = df_texas.groupby('City')['Profit'].sum()
df_tx
```

```python
plt.figure(figsize=(25,25))
plt.rcParams.update({'font.size': 22})
df_store.plot.bar(title="Profit by State", xlabel="State", ylabel="Total Profit")
```

```python
plt.figure(figsize=(15,10))
plt.rcParams.update({'font.size': 15})
df_tx.plot.bar(title="Profit by Each City in Texas", xlabel="City", ylabel="Total Profit", color='crimson')
```
