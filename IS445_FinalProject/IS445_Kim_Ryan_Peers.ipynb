**This is a chart showing the points per game distribution from 2021-2022 season, showing points per gaame based on shooting efficiency for all players active during the season, labeled on a scatter plot.**

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import panel as pn
import numpy as np
```

```python
df = pd.read_csv('all_seasons.csv')
df.head()
```

```python
df_2022 = df[df['season'] == '2021-22']
df_2022.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
x = df['ts_pct']
y = df['pts']
sns.scatterplot(x, y)
plt.xlabel("Shooting Efficiency")
plt.ylabel("Points per Game")
plt.show()
```
