**This is a bar chart showing Dirk Nowitzki's points per game since his debut to the end of his career.**

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import panel as pn
import numpy as np
```

```python
df = pd.read_csv('all_seasons.csv')
df.head()
```

```python
df_dirk = df[df['player_name'] == 'Dirk Nowitzki']
df_dirk.head()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
x = df_dirk['season']
y = df_dirk['pts']
plt.bar(x, y)
plt.title('Dirk Nowitzki Points Per Game per Season')
plt.xlabel("Season")
plt.ylabel("Points per Game")
plt.show()
```
