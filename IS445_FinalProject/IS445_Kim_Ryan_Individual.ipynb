**The "all_seasons.csv" dataset is officially titled "NBA Players", showing statistical information on all NBA players from over two decades, starting from the 1996-1997 NBA season. The dataset contains both numeric and categorical values, with the former containing player information on age, height, weight, points, rebounds, etc., while the latter pieces of data is the the player's name, the last team the player played for, the college they played for (or lack thereof), the season they played in, and their nationality.**

**The dataset appears to be functional, although there is a column that specifically shows the index of each individual player and season stats. **

**This dataset was found in Kaggle using the following link: https://www.kaggle.com/datasets/justinas/nba-players-data**

**The dataset, according to the author, could be used to explore how age/height/weight tendencies have changed over time due to changes in game philosophy and player development strategies, or see how geographically diverse the NBA is and how overseas talents have influenced it.**

**The dataset  itself is 1.75MB compressed in a .zip file.**

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import panel as pn
import numpy as np
```

```python
df = pd.read_csv('all_seasons.csv')
df.head()
```

```python
df_dirk = df[df['player_name'] == 'Dirk Nowitzki']
df_dirk.head()
```

```python
df_dirk['pts'].hist(density=True)
plt.xlabel('Points Per Game')
plt.ylabel('Frequency')
plt.title('Dirk Nowitzkis points per game distribution')
```
